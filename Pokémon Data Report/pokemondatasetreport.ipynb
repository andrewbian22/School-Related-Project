```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import altair as alt
```

```python
df = pd.read_csv('/content/drive/MyDrive/12th Grade/Machine Learning/Data/pokemons.csv')
```

```python
df
```

```python
df['rank']
```

```python
sns.scatterplot(data=df, x='atk', y='def')
plt.title('Attack vs. Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='type1', palette='Set2')
plt.title('Pokemon Types (Type 1)')
plt.xlabel('Type 1')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='type2', palette='Set2')
plt.title('Pokemon Types (Type 2)')
plt.xlabel('Type 2')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='generation', y='total', palette='viridis')
plt.title('Total Stats by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='height', bins=30, kde=True)
plt.title('Distribution of Pokemon Heights')
plt.xlabel('Height (meters)')
plt.ylabel('Count')
plt.show()
```

```python
scatter_plots = []
for pokemon_type in df['type1'].unique():
    filtered_data = df[df['type1'] == pokemon_type]

    scatter_plot = alt.Chart(filtered_data).mark_circle().encode(
        x='atk:Q',
        y='def:Q',
        text='name',
        tooltip=['name', 'atk', 'def']
    ).properties(
        title=f'{pokemon_type} Type Pokemon - Attack vs. Defense'
    )

    scatter_plots.append(scatter_plot)

combined_chart = alt.concat(*scatter_plots, columns=2).configure_title(fontSize=15)

combined_chart.configure_mark(
    fontSize=10,
    font='Helvetica'
)
```
