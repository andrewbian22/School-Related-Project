```python
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')

df.dropna(subset=['latitude'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
plt.figure(figsize=(12, 6))
df['datetime'].dt.year.value_counts().sort_index().plot(kind='line')
plt.title('Number of UFO Sightings Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.grid()
plt.show()

```

```python
# Most common UFO shapes
plt.figure(figsize=(12, 6))
df['shape'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Most Common UFO Shapes')
plt.xlabel('Shape')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

#  UFO sightings by country
plt.figure(figsize=(12, 6))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 5 Countries with Most UFO Sightings')
plt.xlabel('Country (country code)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
# UFO sighting duration distribution
df['duration (seconds)'] = pd.to_numeric(df['duration (seconds)'], errors='coerce')
df = df.dropna(subset=['duration (seconds)'])

plt.figure(figsize=(12, 6))
plt.hist(df['duration (seconds)'], bins=100, edgecolor='k')
plt.title('Distribution of UFO Sighting Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.xlim(0, df['duration (seconds)'].max())
plt.grid(axis='y')
plt.show()
```

```python

#  UFO sightings on a map, lol
plt.figure(figsize=(12, 12))
plt.scatter(df['longitude'], df['latitude'], alpha=0.2, s=10)
plt.title('UFO Sightings Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
```

```python
df['duration (seconds)'] = pd.to_numeric(df['duration (seconds)'], errors='ignore')

df = df[df['duration (seconds)'] < 10000]
plt.figure(figsize=(12, 6))
plt.hist(df['duration (seconds)'].dropna(), bins=50, edgecolor='k')
plt.title('Distribution of UFO Sighting Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

```
