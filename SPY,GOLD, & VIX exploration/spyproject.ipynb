```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert 'Date' column to datetime format
spy['Date'] = pd.to_datetime(spy['Date'])
vix['Date'] = pd.to_datetime(vix['Date'])

# Merge datasets on 'Date'
merged_data = pd.merge(spy, vix, on='Date', how='inner')

# Plotting

# Line plot of SPY Close prices
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['Close'], label='SPY Close', color='blue')
plt.xlabel('Date')
plt.ylabel('SPY Close Price')
plt.title('SPY Close Price Over Time')
plt.legend()
plt.show()

# Line plot of VIX Close prices
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['VIX Close'], label='VIX Close', color='red')
plt.xlabel('Date')
plt.ylabel('VIX Close Price')
plt.title('VIX Close Price Over Time')
plt.legend()
plt.show()

# Scatter plot of SPY Close prices against VIX Close prices
plt.figure(figsize=(12, 6))
plt.scatter(merged_data['Daily Percent Change'], merged_data['VIX Close'], color='green')
plt.xlabel('Daily Percent Change')
plt.ylabel('VIX Close Price')
plt.title('Scatter Plot of SPY Daily Percent Change vs VIX Close')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the three datasets loaded into DataFrames


# Convert 'Date' column to datetime format
spy['Date'] = pd.to_datetime(spy['Date'])
vix['Date'] = pd.to_datetime(vix['Date'])
au['Date'] = pd.to_datetime(au['Date'])

# Merge datasets on 'Date'
merged_data = pd.merge(spy, vix, on='Date', how='inner')
merged_data = pd.merge(merged_data, au, on='Date', how='inner')

# Pairplot to visualize relationships
sns.pairplot(merged_data[['Daily Percent Change_x', 'Daily Percent Change_y', 'Daily Percent Change']], kind='scatter')
plt.suptitle('Pairplot of Daily Percent Changes', y=1.02)
plt.show()

# Correlation matrix
correlation_matrix = merged_data[['Daily Percent Change_x', 'Daily Percent Change_y', 'Daily Percent Change']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

Look at the relationship between Gold and the spy, are they inversely related use the vix as an extension of feature columns for the spy

Did a grid search and combined gold and spy datasets, also tested SVR and Linear Regression model to test which is better and find test correleation

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np


# Merge datasets on 'Date' column
merged_data = pd.merge(spy, au, on='Date', suffixes=('_spy', '_gold'))

# Select relevant columns
X = merged_data[['Open_spy', 'High_spy', 'Low_spy', 'Open_gold', 'High_gold', 'Low_gold']]
y_spy = merged_data['Close_spy']
y_gold = merged_data['Close_gold']

# Train-test split
X_train, X_test, y_spy_train, y_spy_test, y_gold_train, y_gold_test = train_test_split(
    X, y_spy, y_gold, test_size=0.2, random_state=42
)

# RandomForestRegressor parameter grid
param_dist = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Randomized search for RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)
rf_random = RandomizedSearchCV(rf_model, param_distributions=param_dist, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rf_random.fit(X_train, y_spy_train)

# Best parameters for RandomForestRegressor
best_rf_params = rf_random.best_params_

# Train RandomForestRegressor with best parameters
rf_best_model = RandomForestRegressor(**best_rf_params, random_state=42)
rf_best_model.fit(X_train, y_spy_train)

# Other models for comparison
models = {
    'Linear Regression': LinearRegression(),
    'SVR': SVR(),
    # Add more models if needed
}

# Train and evaluate other models
for name, model in models.items():
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(X_train, y_spy_train)
    y_pred = pipeline.predict(X_test)
    mse = mean_squared_error(y_spy_test, y_pred)
    print(f'{name} MSE: {mse}')

# Find the strongest correlation
correlation_spy_gold = np.corrcoef(y_spy_test, y_gold_test)[0, 1]
print(f'Correlation between SPY and Gold: {correlation_spy_gold}')

# Additional analysis or plotting as needed
```
